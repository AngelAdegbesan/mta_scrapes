```python

import os ## for file/folder management
from pathlib import Path ## to provide a path to files/folders
## import the glob library for collecting specific files into a list
import glob
import fileinput
import pandas as pd ## to export csv file
```

```python
os.listdir()
```

```python
## let's capture the files in a list
txt_files = sorted( glob.glob("*.txt"))
txt_files
```

```python
## let's capture the 2021 files in a list
txt21_files = sorted( glob.glob("turnstile_21*.txt"))
txt21_files
```

```python
## let's capture the 2022 files in a list
txt22_files = sorted( glob.glob("turnstile_22*.txt"))
txt22_files
```

```python
len (txt22_files)
```

```python
#working with the 2021 files
header = "C/A,UNIT,SCP,STATION,LINENAME,DIVISION,DATE,TIME,DESC,ENTRIES,EXITS"
stations21 = []
date21 = []
entries21 = []
exits21 = []
```

```python
##extracting data for the data dictionary
for line in fileinput.input(txt21_files):
    if line.startswith(header):
        continue
    data = line.split(",")
    stations21.append(data[3])
    date21.append(data[6])
    entries21.append(int(data[9]))
    exits21.append(int(data[10]))
```

```python
data_dict21 = {
 "station": stations21,
 "date": date21,
 "entries": entries21,
 "exits": exits21
}
data_frame21 = pd.DataFrame(data_dict21)
data_frame21["date"] = pd.to_datetime(data_frame21["date"])
data_frame21["entries"] = pd.to_numeric(data_frame21["entries"])
data_frame21["exits"] = pd.to_numeric(data_frame21["exits"])
```

```python
len(data_frame21)
```

```python
stations21 = data_frame21.station.unique().tolist()
```

```python
grouped_data21 = {}
for station in stations21:
    df = data_frame21.loc[data_frame21["station"] == station]
    grouped_data21[station] = df
```

```python
grouped_data21_by_months = {}
for station in stations21:
    df = grouped_data21[station]
    january21_df = df.loc[df["date"].dt.month == 1]
    february21_df = df.loc[df["date"].dt.month == 2]
    march21_df = df.loc[df["date"].dt.month == 3]
    april21_df = df.loc[df["date"].dt.month == 4]
    grouped_data21_by_months[station] = {
        "january": january21_df,
        "february": february21_df,
        "march": march21_df,
        "april": april21_df
    }
```

```python
grouped_data21_by_months[stations21[0]]["february"]
```

```python
total21_by_months = {}
for station in stations21:
    print(f"On station {station} -> {stations21.index(station)} of {len(stations21)}")
    total21_by_months[station] = {}
    for month in grouped_data21_by_months[station]:
        df = grouped_data21_by_months[station][month]
        entries21 = df["entries"].sum()
        exits21 = df["exits"].sum()
        total21_by_months[station][month] = {
            "entries": entries21,
            "exits": exits21
        }
```
